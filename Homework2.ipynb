```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q1-Read the data https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv

data = "https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv"
df = pd.read_csv(data)
df.head()
```

```python
#Q2- Get data information and number of rows and columns
df.info()
print("\nShape of the dataset (Rows, Columns):")
print(df.shape)
```

```python
#Q3-Get first 3 rows of data
df.head(3)
```

```python
#Q4-Get descriptive statistics of data
df.describe()
```

```python
#Q5-Use seasoborns heat map to generate following heatmap
# annot = True to print the correlation inside the square

corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0)
plt.show()
```

```python
#Q6-Plot the density distribution of Balance

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Balance', fill=True, color='blue')

plt.xlabel('Balance')
plt.ylabel('Density')

plt.show()
```

```python
#Q7-Count the frequency values of each Region

region_counts = df['Region'].value_counts()
print("Frequency of each Region:")
print(region_counts)
```

```python
#Q8-List Percentage of  the Married people

married_percentage = df['Married'].value_counts(normalize=True) * 100
print("\nPercentage of Married people:")
print(married_percentage)
```

```python
#Q9-Create histogram of Balance

plt.figure(figsize=(10, 6))
sns.histplot(df['Balance'], bins=20, kde=False, color='skyblue', edgecolor='black')

plt.title('Histogram of Balance')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q10-Get the frequency distribution of Region using bar graph

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Region', palette='viridis')

plt.title('Frequency Distribution of Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
#Q11-Get the mean income  of married people?

mean_income_married = df[df['Married'] == 'Yes']['Income'].mean()

print(f"The mean income of married people is: {mean_income_married}")
```

```python
#Q12-Get the scatter plot of Income and Limit

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='Limit')

plt.title('Scatter Plot of Income vs Limit')
plt.xlabel('Income')
plt.ylabel('Limit')
plt.show()
```

```python
#Q13-Create new column named IncomebyLimit by dividing Income to Limit.

df['IncomebyLimit'] = df['Income'] / df['Limit']
df[['Income', 'Limit', 'IncomebyLimit']].head()
```

```python
# Q14-Create the pair plot using Seaborn library's pairplot

sns.pairplot(df)
plt.show()
```

```python
#Q15- drop new column named IncomebyLimit from dataframe

df.drop(columns=['IncomebyLimit'], inplace=True)
print("Columns total:")
print(df.columns)
```

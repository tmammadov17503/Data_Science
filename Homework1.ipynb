```python
#Q8-Write a class named arithmetic with method for calculating addition, substraction, multiplication and division of two numbers.
class arithmetic:
  def addition(x,y):
    return x+y
  def subtraction(x, y):
    return x - y
  def multiplication(x, y):
    return x * y
  def division(x, y):
    if y == 0:
      return "Error: Division by zero is not allowed."
    return x / y

print(f"Addition: {arithmetic.addition(3, 4)}")
print(f"Subtraction: {arithmetic.subtraction(10, 5)}")
print(f"Multiplication: {arithmetic.multiplication(6, 7)}")
print(f"Division: {arithmetic.division(20, 4)}")
```

```python
#Q9- write a for loop or method  to calculate summation of first n numbers
#i.e if n=3, it will print or return 6 (1+2+3)

def sum_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

n = 3
print(sum_numbers(n))
```

```python
#Q10-Generate 10 numbers between 0 and 1 and add into list. Print first three elements of that list.

import random

random_list = []
for i in range(10):
    random_list.append(random.random())

print("Full List:", random_list)
print("First three elements:", random_list[:3])
```
